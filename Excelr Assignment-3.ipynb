```python
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv('sales_data_with_discounts.csv')
```

```python
# changing the datatype of date column from object to datetime
df['Date']=df['Date'].astype('datetime64[ns]')
```

```python
df.info()
```

```python
df.describe(include=[int,float])
```

```python
numerical_columns=df.select_dtypes(include=[int,float]).columns
numerical_columns
```

```python
for i in numerical_columns:
    plt.figure(figsize=(6,3))
    sns.histplot(df[i],kde=True,color='red')
    plt.show()
```

```python
for i in numerical_columns:
    print(f'{i}:{round(df[i].skew(),2)}')
    
# this shows that columns(volume, avg price,total sales value,discount amount,net sales value) are highly right skewed and column (Discount rate) is left skewed
```

```python
for i in numerical_columns:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df[i])
    plt.show()
```

```python
categorical_columns=df.select_dtypes(include='object').columns
categorical_columns
```

```python
for i in categorical_columns:
    plt.figure(figsize=(6,3))
    sns.countplot(x=df[i],hue=df[i])
    plt.tight_layout()
    plt.xticks(rotation=90)
    plt.show()
    
```
