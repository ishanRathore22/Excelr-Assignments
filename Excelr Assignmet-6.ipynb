a. Build 99% Confidence Interval Using Sample Standard Deviation
Assuming the sample is representative of the population, construct a 99% confidence interval for the mean number of characters printed before the print-head fails using the sample standard deviation. Explain the steps you take and the rationale behind using the t-distribution for this task.


```python
import numpy as np
data=np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
```

```python
from scipy import stats
```

```python
loc=data.mean()
n=len(data)
df=n-1
scale=data.std()/np.sqrt(n)
lower_limit,upper_limit=stats.t.interval(0.99,df=df,loc=loc,scale=scale)
```

```python
print('Lower Limit',round(lower_limit,3))
print('Upper Limit:',round(upper_limit,3))
```

"we use t distribution for this task becaue there is no information about population is given and when information about sample is given we mostly go for t distribution table" 

b. Build 99% Confidence Interval Using Known Population Standard Deviation
If it were known that the population standard deviation is 0.2 million characters, construct a 99% confidence interval for the mean number of characters printed before failure.


```python
confidence=0.99
pop_std=0.2
loc=data.mean()
n=len(data)
scale=pop_std/np.sqrt(n)
```

```python
lower_limit,upper_limit=stats.norm.interval(confidence,loc=loc,scale=scale)
```

```python
print('Lower Limit',round(lower_limit,3))
print('Upper Limit:',round(upper_limit,3))
```

# Hypothesis Testing

```python
sample_mean=3050
x=600
population_mean=1000+5*x
population_std=5*25 
sample_size=25
alpha=0.05
```

```python
# step 1 - state null and alternate Hypothesis
h0="weekly operating cost are 4000"
h1="weekly operating cost are greater than 4000"
```

```python
# step 2 - find z value
SE=population_std/np.sqrt(sample_size)
z_value=(sample_mean-population_mean)/SE
z_value
```

```python
# step 3 find p value through z value
p_value=1-stats.norm.cdf(z_value)
```

```python
# step 4 check if p value is greater or less than critical value
if p_value>alpha:
    print('We Failed to Reject null Hypothesis')
else:
    print('We failed to Accept null Hypothesis')
```
